```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

import skimage
from skimage import io
```

```python
rootRadius = 26;
besselRadius = sp.special.jn_zeros(0, rootRadius)[rootRadius - 1];

resolution = 2047;
pixelRadius = resolution / 2;
```

```python
def EvaluateBessel(x, y):
    r = np.sqrt(x**2 + y**2)
    return sp.special.j0(r / resolution * besselRadius)
```

```python
besselMap = np.fromfunction(lambda x, y: EvaluateBessel((x + 0.5 - pixelRadius), (y + 0.5 - pixelRadius)), (resolution, resolution), dtype = np.single)
```

```python
plt.imshow(besselMap)
```

```python
io.imsave("BesselMap.png", besselMap * 0.5 + 0.5)
```
